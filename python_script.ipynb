```python
#Import our Libraries
import pandas as pd
```

```python
#Read in our CSV
df=pd.read_csv("customer_orders.csv")

df
```

```python
df.columns
```

```python
df_uniq=df[['order number', 'employee id', 'employee name', 
       'employee_job_title', 'order date', 'order type', 'customer type',
       'cust id', 'customer name', 'prod category',
       'prod number', 'prod_name', 'quantity', 'price', 'order total']]
```

```python
df_uniq
```

```python
df_nosp_unique=df_uniq.reset_index()

df_nosp_unique.columns=df_nosp_unique.columns.str.replace(' ','')
```

```python
df_nosp_unique
```

```python
df_nosp_unique.head()
```

```python
df_nosp_unique[df.duplicated()]
df_no_dup=df_nosp_unique.drop_duplicates(subset=['ordernumber','employeeid'])
```

```python
df_no_dup.head()
```

```python
import sqlite3


conn1=sqlite3.connect('database.db')
df_no_dup.to_sql('order_table',conn1,if_exists='replace',index='False')
```
